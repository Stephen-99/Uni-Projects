## Abstract
This report analyses the brine shrimp simulation. The parameters being investigated are, chance of death, number of shrimp, screen size and time between stages. 

## Background
The simulation emulates the behaviour of brine shrimp. It does this by simulating their movement and and interaction within a confined tank. The shrimp bounce of each other and the walls. Any collisions between male and female shrimp results in a couple. They then have a chance to reproduce and create new eggs.

The chance of death was chosen as a parameter to investigate, as it will clearly impact the chance of survival of the shrimp and therefore, the fluctuation of shrimp over time. The initial number of shrimp was also deemed suitable since the higher the initial number, the more likely collisions will occur and therefore the more likely reproduction will occur and the population will grow. The screen size was chosen since a smaller screen size forces the shrimp closer together and should increase the number of collisions.


## Methodology
The simulations have been compared using parameter sweeps. The first parameter sweep iterated the chance of death multiplier from 0 to 2 with a step size of 0.4. The Number of shrimp was looped from 20 to 100 with a step size of 20. the screen size was incremented with the x and y dimensions simultaneously. The x dimension increased from 600 to 1600 in increments of 200. The y dimension incremented from 400 to 950 in steps of 110. The code to run this is ```bash paramSweep.sh```

Initially this sweep was performed with the chance for death multiplier multiplying the input to the cubic function rather then the output of the cubic function. This moved the chance of death along the graph and while changing the chance of death, it was not necessarily increasing or decreasing it as expected.As can be seen from the graphs below where the time to reach 500 shrimp decreases with a higher multiplier (0.4 to 0.8). If the multiplier was increasing the chance for a shrimp to die as it should be, then we would expect it to be harder for the population to increase and so take a longer time. There is of course random chance involved in the calculations and so the results may vary.

## Results



The varying screen size had an unpredicted effect. Small screen sizes did not increase population faster but rather increased time to reach 500 shrimp. Increasing the size tended to increase rate of growth then decrease it again. An explanation for this is that the smaller space is forcing shrimp closer together and creating increased collisions. While this is helps increasing the number of couples, it also causes eggs to collide and annihilate each other. It is also noteworthy that when two shrimp form a couple, the shrimp count will decrease since a couple is stored as an individual shrimp. An increase in collisions may lead to excessive collisions so that almost no individual adults are present but are instead present as couples. 
This lead to circumstances where the shrimp count fluctuated around the same value, staying fairly stable. See an example below:

The number of shrimp also had a notable impact on the rate of increase of the population. As the number of initial shrimp increased, there where more who survived to be an adult. The time to change affected the results in an unexpected way. Initially, it was expected that the time to change would not affect results since the chance for death is normalized by time 2 change. However, a longer time to change clearly means more iterations before changing stage even if the chance of death is the same. A longer time to change therefore results in less shrimps making it to the adult stage where they are able to reproduce. As a result, increasing the time to change significantly decreases overall chances of survival. Additionally, it reduces rate of population growth.

The images below demonstrate this, with time to increase changing from left to right and constant population, dimensions and chance of death are used. Their values are 100, 1400 x 900 and 1.3 respectively. The number of iterations to achieve a population of 500 increases, with the following graphs demonstrating the decreasing number of iterations required to reach 0 population.

## Conclusion and Future Work

It is clear that the number of number of shrimp, screen dimensions, chance of death and time to change all had an impact on the results.
The chance of survival increased with the number of shrimp and decreases with an increase in chance of death and time to change. The screen dimensions had a interesting effect, small screen sizes squash shrimp together and can kill new eggs, with larger screens alleviating this problem. Once the problem is no longer prevalent, increasing screen size decreases chance of survival.

To extend on this analysis, repeat trials could be conducted and the averages analysed achieve more reliable results. It would also be beneficial to store the number of shrimp in a file so that tables of results could be constructed.
